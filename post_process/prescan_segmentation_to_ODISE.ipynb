```python
import numpy as np
from PIL import Image
import os

# Define the neutral color and thresholds for other colors
neutral_color = (84, 1, 68) 
thresholds = {
    'sky': ([128, 4, 252], 80, neutral_color),
    'car': ([255, 255, 255], 80, (71, 192, 110)),
    'truck': ([1, 255, 1], 80, (71, 192, 110)),
    'road_markings': ([255, 255, 0], 80, (255, 255, 255)),
    'human': ([0, 0, 255], 80, (71, 30, 112)),
    'road': ([126, 126, 126], 80, (255, 255, 255)),
    'tree' : ([128,4,4], 80, (128,4,4))
}

# Function to map RGB colors
def map_rgb(image, thresholds, neutral_color):
    img_array = np.array(image)
    
    # Initialize the mask with the neutral color
    mapped_img_array = np.full(img_array.shape, neutral_color, dtype=np.uint8)

    for key, (target_color, tolerance, new_color) in thresholds.items():
        mask = np.all(np.abs(img_array - target_color) <= tolerance, axis=-1)
        mapped_img_array[mask] = new_color

    return Image.fromarray(mapped_img_array)


# dirs = ["stop", "straight"]
# for dir in dirs:
#     # Get all the images in the segmentation folder
#     base_path = f"/mnt/d/marketing/process/0241/{dir}/"
#     path = os.path.join(base_path, "seg_70")
#     output_path = os.path.join(base_path, "seg_ODISE")

#     for root, dirs, files in os.walk(path):
#         for file in files:
#             if file.endswith(".png"):
#                 dir_name = root.split("/")[-1]
#                 im = Image.open(os.path.join(root, file))

#                 # Save the changed image
#                 destination = output_path
#                 if not os.path.exists(destination):
#                     os.makedirs(destination)

#                 output_file_path = os.path.join(destination, file)
#                 if os.path.exists(output_file_path):
#                     continue

#                 # Map RGB colors efficiently
#                 mapped_image = map_rgb(im, thresholds, neutral_color)

#                 # Save the mapped image
#                 mapped_image.save(output_file_path)

#                 # Close the image file
#                 im.close()


im = Image.open("/home/samuele/dev/Ctrl-Adapter/assets/evaluation/frames/segmentation/nuscenes_scene_0001/ImageSegmentationSensor_1_00000.png")



# Map RGB colors efficiently
mapped_image = map_rgb(im, thresholds, neutral_color)

# Save the mapped image
mapped_image.save("test_ODISE.png")

# Close the image file
im.close()
```
